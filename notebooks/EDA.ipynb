```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../data/processed/listings_cleaned.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))

df.head()
```

```python
df.info()
```

```python
# missing value
missing = df.isnull().sum()

missing = missing[missing > 0]

missing.sort_values(ascending=False)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["price_clean"],
    bins=50
)

plt.title("Distribution of Airbnb Prices")
plt.xlabel("Price")
plt.show()
```

```python
top10 = (
    df["neighbourhood_cleansed"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top10.values,
    y=top10.index
)

plt.title("Top 10 Neighbourhoods")
plt.show()
```

```python
room_counts = df["room_type"].value_counts()

plt.figure(figsize=(8,6))

plt.pie(
    room_counts,
    labels=room_counts.index,
    autopct="%1.1f%%"
)

plt.title("Room Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x="room_type",
    y="price_clean",
    data=df
)

plt.xticks(rotation=45)

plt.show()
```

```python
cols = [
    "price_clean",
    "accommodates",
    "bedrooms",
    "beds",
    "bathrooms",
    "number_of_reviews",
    "review_scores_rating"
]

corr = df[cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    x="accommodates",
    y="price_clean",
    data=df
)

plt.show()
```
